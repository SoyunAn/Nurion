```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data_4_64 = pd.read_csv('4_64.csv',sep=',',index_col=None)
data_8_64 = pd.read_csv('4_128.csv',sep=',',index_col=None)
data_16_64 = pd.read_csv('4_256.csv',sep=',',index_col=None)
data_64_64 = pd.read_csv('4_512.csv',sep=',',index_col=None)
```

```python
x1=data_4_64['step']
y1=data_4_64['time']

x2=data_8_64['step']
y2=data_8_64['time']

x3=data_16_64['step']
y3=data_16_64['time']

x4=data_64_64['step']
y4=data_64_64['time']
```

```python
plt.figure(figsize=(20,20))
plt.plot(x1,y1,'or')
plt.plot(x2,y2,'og')
plt.plot(x3,y3,'ob')
plt.plot(x4,y4,'ok')
plt.xlabel('global step',fontsize=30)
plt.ylabel('time',fontsize=30)
#plt.ylim(0,10)
plt.grid()

plt.legend(("batch64","batch128","batch256","batch512"),fontsize=30,loc='upper right')
#plt.legend(("CPU4","CPU8","CPU16","CPU64"),fontsize=30,loc='upper right')

#plt.yscale('log')
#plt.show()
plt.savefig("time_step2.jpg")
```
