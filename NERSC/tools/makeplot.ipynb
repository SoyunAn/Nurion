```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folder='result_64nd/'
path=folder+'out.txt'

cnt=0

#import sys
#sys.stdout = open('4node_sum.txt','w')

loss=[]
val_loss=[]


with open(path) as f:
    for line in f:
        line=line.strip()
        #if 'val_acc' in line or 'Epoch ' in line and 'improve' not in line:         
        if 'val_acc' in line:
            loss.append(line.split()[7])
            val_loss.append(line.split()[13])

df = pd.DataFrame({'loss':loss,'val_loss':val_loss},index=range(1,65))

plt.figure(figsize=(12, 12))

#plt.plot(df['loss'],'bo',label='Traning loss')
plt.plot(df['val_loss'],'bo',label='Validation loss')
plt.title('Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
ax = plt.gca()
#ax.set_ylim([0.12])
ax.invert_yaxis()
plt.savefig("64node_valloss.png")
print(df)
```
